```python
def mult_without_comprehension(num):
    for n in num:
        n *= 10

    return num
```

```python
def mult_with_comprehension(num):
    return [n * 10 for n in num]
```

```python
%%timeit

mult_without_comprehension(range(1000))
```

```python
%%timeit

mult_with_comprehension(range(1000))
```

```python
%%timeit

mult_without_comprehension(range(10_000))
```

```python
%%timeit

mult_with_comprehension(range(10_000))
```

```python
import time

n_trials = 1000

result = []

for _ in range(n_trials):
    t_0 = time.time()
    mult_without_comprehension(range(10_000))
    t_elapsed = time.time() - t_0
    result.append(t_elapsed)

print(result)
print(sum(result) / len(result))
```
